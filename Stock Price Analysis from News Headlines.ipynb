```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\VBALA\Downloads\Combined_News_DJIA.csv",encoding="ISO-8859-1")
```

```python
df.head(10)
```

```python
train=df[df['Date']<'20150101']
test=df[df['Date']<'20141231']
```

```python
#remove special characters
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
#Renaming columns
list1=[i for i in range(25)]
newind=[str(i) for i in list1]
data.columns=newind
data.head(5)
```

```python
#Convert text to lower case
for i in newind:
    data[i]=data[i].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines=[]
for i in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[i,0:25]))
print(headlines[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
# BAG OF WORDS
tfidfvector=TfidfVectorizer(ngram_range=(2,2))
#all combinations of pair of words will be vectorized
traindataset=tfidfvector.fit_transform(headlines)
print(traindataset)
```

```python
#RandomForest Classifier
randomclass=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclass.fit(traindataset,train['Label'])
```

```python
#Test Data Set
test_transform=[]
for i in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[i,2:27]))
test_dataset=tfidfvector.transform(test_transform)
predictions=randomclass.predict(test_dataset)
print(predictions.shape)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix) #inidicates true positive and true negative
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
